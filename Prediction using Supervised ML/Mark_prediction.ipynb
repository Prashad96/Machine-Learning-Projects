# Prediction using Supervised ML
## Predicting the percentage of an student based on No. of study hours.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('data.csv')
```

```python
x = data['Hours']
y = data['Scores']
y = y[:,np.newaxis]
x = x[:,np.newaxis]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=.2)
```

```python
# plotting train data
plt.title("Visualizing training data")
plt.xlabel('Study hours')
plt.ylabel('Marks obtained')
plt.scatter(X_train,y_train)
```

```python
# Training the model
model = LinearRegression().fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Visualizing the predictions
plt.title("Visualizing the trained model")
plt.xlabel('Study hours')
plt.ylabel('Marks obtained')
plt.scatter(X_train,y_train,c='r')
plt.plot(X_train,model.predict(X_train))
```

```python
plt.title("Visualizing the Predictions")
plt.xlabel('Study hours')
plt.ylabel('Marks obtained')
plt.scatter(X_test,y_test,c='r')
plt.plot(X_train,model.predict(X_train))
```

```python
print('Mean Squared Error : ',mean_squared_error(y_test,y_pred,squared=False))
```
